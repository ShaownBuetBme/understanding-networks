### Train the model

```python
from ultralytics import YOLO

# Load pretrained detection model (not segmentation)
model = YOLO('yolov8n.pt')  # you can also use yolov8s.pt or yolov8m.pt depending on your needs

# Train
model.train(
    data='/content/datasets/Chessboard-3/data.yaml',
    epochs=10,
    imgsz=640,
    batch=16,
    project='YOLO_weights',
    name='Checkerboard_detection_output',
    exist_ok=True
)

```

## Model test (Detection)

```python
from ultralytics import YOLO
import cv2
from matplotlib import pyplot as plt

# Load your trained model (best.pt after training)
model = YOLO('/content/YOLO_weights/Checkerboard_detection_output/weights/best.pt')

# Path to test image
image_path = '/content/datasets/Chessboard-3/test/images/20_f_jpg.rf.53f26510b333e7432d12d32ab2ee1c54.jpg'

# Run inference
results = model(image_path, conf=0.5)

# Access first result (because results is a list)
result = results[0]

# Show results directly (this works inside notebook)
result.show()

# If you want to visualize inline (better for notebook)
# Load original image
img = cv2.imread(image_path)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Plot image using matplotlib
plt.figure(figsize=(10, 10))
plt.imshow(img)
plt.axis('off')
plt.title('Original Image')
plt.show()

# You can also print detections:
for box in result.boxes:
    cls = int(box.cls[0])
    conf = float(box.conf[0])
    xyxy = box.xyxy[0].cpu().numpy()
    print(f"Class: {cls}, Confidence: {conf:.2f}, Box: {xyxy}")
```
